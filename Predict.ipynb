```python
import pandas as pd
training_data = pd.read_csv('training_data_grouped.csv', header=None, names=["cleaned_chord", "C", "Cs", "D", "Ds", "E", "F", "Fs", "G", "Gs", "A", "As", "B", "count"])
```

```python
training_data
```

```python
ug_chords = ['C', 'D', 'D/F#', 'Em', 'G', '<Silent>']
training_data = training_data.loc[training_data['cleaned_chord'].isin(ug_chords)]
training_inputs = training_data[["C", "Cs", "D", "Ds", "E", "F", "Fs", "G", "Gs", "A", "As", "B"]]
training_outputs = training_data.cleaned_chord
training_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
n_neighbors_size = min(len(training_inputs.index), len(ug_chords))
knn_classifier = KNeighborsClassifier(n_neighbors=1)
knn_classifier.fit(training_inputs, training_outputs)
```

```python
chromagram = pd.read_csv('chromagram.csv')
```

```python
COL_NAMES_NOTES = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
predictions = knn_classifier.predict(chromagram[COL_NAMES_NOTES])
chromagram['predicted'] = predictions
                
```
